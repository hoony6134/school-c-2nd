## Imports



````python
# tensorflow, keras, numpy 라이브러리 사용 선언
import tensorflow as tf
import numpy as np
from tensorflow import keras
````

## 전체코드

````python
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer = 'sgd', loss = 'mean_squared_error')
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype = float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype = float)
model.fit(xs, ys, epochs=500)
model.predict([10])
````

# 전체 코드 뜯어보기

## Define and Compile the Neural Network

1개의 레이어, 그리고 1개의 뉴런을 가지고 있는 뉴럴 네트워크 모델링 하기

````python
# 인공지능 모델 정의: 1 layer, 1 neuron, 1 output
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
# 인수(매개변수)에 대한 설명 : https://keras.io/api/layers/core_layers/ (Dense: https://keras.io/api/layers/core_layers/dense/)
# Sequential: 계층이 순서대로 정의된다는 의미
# layer: 층
# Dense: 밀도, 연결된 뉴런의 층을 정의하기 위해 Dense라는 단어 사용
# units: 뉴런의 갯수
# input_shape: 입력(input)데이터의 차원을 정의하여 어떤 형태로 입력되는지 나타냄(일반적으로 numpy의 array형태로 입력되어 사용함)
````

컴파일 하기 위해서는 2개의 함수가 필요하다. 

loss 와 optimizer.

학습을 통해 예측한 값과, 실제의 y=10x+10 값 두개를 측정하는 LOSS function 을 설정해야 한다.

위의 로스를 최소화 하기 위해서 OPTIMIZER function 을 설정한다.

로스함수로 'MEAN SQUARED ERROR'를 사용하고, 옵티마이저로 'STOCHASTIC GRADIENT DESCENT' 를 사용한다. 아직 수학적으로 알 필요는 없다.


````python
# 모델을 정의하고 optimizer(최적화), loss function(손실 함수) 설정
model.compile(optimizer = 'sgd', loss = 'mean_squared_error')
# optimizer & 수학 및 이론적 설명: https://medium.com/analytics-vidhya/optimization-algorithms-for-deep-learning-1f1a2bd4c46b
# 학습(taining) 시 오류(손실) 수정이 포함되므로 loss와 optimizer가 필요하며 컴파일을 해야 함
# loss function: 머신러닝을 통해 규칙을 만드는데, 예측이 얼마나 잘 됐는지 평가해야 의사 결정이 가능하므로, loss function은 이를 위한 함수
# loss가 적을수록 '수렴' 된다고 표현함
# mean_squared_error 평균제곱오차: 실제값과 예측값의 차이를 계산함
# optimizer: loss function의 데이터를 사용해 추측이 얼마나 잘 수행되었는지를 판단하여 모델을 최적화하기 위해 어떻게 할 것인지 나타냄
# sgd로 측정하고 데이터를 받아 다음 추측값을 찾음
# sgd(Stochastic gradient descent) 확률적 경사 하강방식
````

## Providing the Data

y=2x-1 이 식에 따라, x의 데이터 6개와, 그에 해당하는 y의 데이터 6개를 준비한다. 머신러닝은 데이터로부터 학습하는 것이기 때문이다. 이 데이터로부터 우리는 위의 식을 컴퓨터가 찾아내도록 하는것이다. 

'Numpy' 라이브러리를 이용해서, xs 와 ys 를 만든다. 각각 6개씩

````python
# 훈련세트와 테스트 세트에서 사용할 6개의 숫자 배열 생성(학습 데이터 정의)
# x와 y 집합의 관계를 분석을 위함
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype = float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype = float)
````

# Training the Neural Network

x와 y의 관계를 찾기 위해서, 학습하는 함수인 **model.fit** 함수를 호출한다. 각 에포크별로 좋은 모델인지 아닌지 평가할 수 있는 loss 값을 제공한다. 

````python
model.fit(xs, ys, epochs=500)
# xs와 ys 간의 관계를 학습함
# model.fit() 함수: 모델이 기능을 추측하고 손실을 측정하며 최적화를 통해 다른 추측을 반복함 
# epochs = 500 은 훈련을 500번 거친다는 의미
````

모델의 학습이 끝나면 이제 **model.predict** 함수를 이용해서 새로운 데이터를 예측할 수 있다. 10을 넣어보자. 무슨 값이 예측되었는가?

````python
# 모델을 학습시켜 규칙을 만드는 단계
# model.predict([10.0])
model.predict([10])

# 결과값을 출력해봤을 때, 정확히 19가 아니라 18.977009와 같이 애매하게 비슷한 값이 나옴
# 19와 더 근접하게 출력되도록 하기 이해 어떻게 해야 할까? 
# 방법> 제공하는 데이터를 늘리기!
````

19가 정확히 나오지 않은 이유는, 뉴럴넷은 확률로 처리한다. 따라서 주어진 데이터 6개를 보고, 식이 Y=2X-1 과 비슷할 것이라고 예측한 것이다. 그래서 거의 19라고 예측을 해준다.
